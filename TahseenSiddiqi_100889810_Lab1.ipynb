```python
#Importing all the desired packages and modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.models import load_model
```

```python
data=pd.read_csv('hepatitis.data.csv',header=None,na_values='?')
data.head()
```

Since we don't have column names , we will assign it to them

```python
#Applying columns
columns=['Class','AGE','SEX','STEROID0','ANTIVIRALS','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM',
        'SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']
data.columns=columns
```

```python
data.head()
```

```python
# Percentage missing values
def missing_datas(dataset):
    total = dataset.isnull().sum().sort_values(ascending=False)
    percent = (dataset.isnull().sum()/dataset.isnull().count()*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data

missing_data = missing_datas(data)
print(missing_data)
```

Let's replace all the Nan values with mean as asked in the instruction

```python
for i in data:
    data[i]=data[i].replace(np.nan,data[i].mean())
```

```python
#Lets' check for the null values again
data.info()
```

We have sucessfully replaced the nan values with mean.

Before we fit the AN

Before we fit the ANN model in the dataset we will divide the dataset into features and target value

```python
#Diving X,y and train and test dataset
X=data.drop(['Class'],axis=1)
Y=data['Class']
dim=len(X.columns)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 32)
```

```python
#Labelencoding the labels/target values to bring it to 0 and 1 for binary prediction
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
print(y_train,y_test)
```

```python
#Scaling the data to bring the large continous values to same scale as others
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Creating Simple ANN as per the instructions 
model = Sequential()
model.add(Dense(19, input_dim=dim, activation='sigmoid'))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(15, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=250, batch_size=32)
```

```python
loss = model.evaluate(X_test, y_test, verbose=2)
```

```python
#saving on disk
model.save('model.h5')
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
#Setting threshold 0.5 probability
predicted=[1 if x > 0.5 else 0 for x in y_pred]
print(predicted)
```
